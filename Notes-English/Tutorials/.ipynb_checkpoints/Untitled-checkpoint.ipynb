```julia
using Pkg
Pkg.activate("../.")
using OptimalTransport, BenchmarkTools, LinearAlgebra, Distributions, Distances
using Random
using PDMats
using Symbolics
using Plots
using Tulip
using HCubature
```

```julia
μ = MvNormal([0,0],[1 0;0 2])
ν = MvNormal([10,10],[2 0;0 1])
```

```julia
Δ = 0.5
thrs = 1e-5
x = -3:Δ:3
y = -4:Δ:4
f(x,y) = pdf(μ,[x,y])
X = repeat(reshape(x, 1, :), length(y), 1)
Y = repeat(y, 1, length(x))
Z = map(f, X, Y)
Z = Z./sum(Z)
μ_supp = hcat(reshape(X,:,1),reshape(Y,:,1))
μ_n    = hcat(reshape(Z,:,1))

g(x,y) = pdf(ν,[x,y])
x = 6:Δ:14
y = 7.5:Δ:12.5
X = repeat(reshape(x, 1, :), length(y), 1)
Y = repeat(y, 1, length(x))
Z = map(g, X, Y)
Z = Z./sum(Z)
ν_supp = hcat(reshape(X,:,1),reshape(Y,:,1))
ν_m    = hcat(reshape(Z,:,1));

C = pairwise(SqEuclidean(),μ_supp',ν_supp');
```

```julia
θ  = collect(0:0.2:2π);
θx = cos.(θ)
θy = sin.(θ)
δ  = collect(0:0.2:1)
posμ = [0.0 0.0]
posν = [10. 10.]
for i in δ[2:end]
    a = [θx.*i θy.*i*2]
    b = [θx.*i*2 θy.*i] .+ [10 10]
    posμ = vcat(posμ,a)
    posν = vcat(posν,b)
end

μ_s = posμ
ν_s = posν
μ_p = pdf(μ,μ_s')
μ_p = μ_p./sum(μ_p)
ν_p = pdf(ν,ν_s')
ν_p = ν_p./sum(ν_p);
C = pairwise(SqEuclidean(),μ_s',ν_s');
```

```julia
isapprox(emd2(μ_p,ν_p,C,Tulip.Optimizer()),ot_cost(SqEuclidean(),μ,ν),rtol=1e-3)
```

```julia
    T = ot_plan(SqEuclidean(),μ,ν)
```

```julia
f(x) = sqeuclidean(x,T(x))*pdf(μ,x)
isapprox(hcubature(f,[-10,-10],[10,10])[1],ot_cost(SqEuclidean(),μ,ν),rtol=1e-3)
```

```julia

err = emd2(μ_p,ν_p,C,Tulip.Optimizer())-ot_cost(SqEuclidean(),μ,ν)
rerr= err/ot_cost(SqEuclidean(),μ,ν)
```
