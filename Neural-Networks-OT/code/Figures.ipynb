```julia
using Distributions
using CairoMakie, Random
CairoMakie.activate!(type = "svg")



Random.seed!(7)
n = 200

fig = Figure(resolution = (700, 700))
ax1 = Axis(fig,xgridstyle=:dash,ygridstyle=:dash,xtickalign=1, ytickalign=1)
ax2 = Axis(fig,xgridstyle=:dash,ygridstyle=:dash,xtickalign=1, ytickalign=1)
ax3 = Axis(fig,xgridstyle=:dash,ygridstyle=:dash,xtickalign=1, ytickalign=1)
lines!(ax1,-3:0.1:3,pdf(Normal(0,1),[-3:0.1:3])[1])
lines!(ax3,pdf(Normal(0,1),[-3:0.1:3])[1],-3:0.1:3.)

xs = LinRange(-4, 4, 100)
ys = LinRange(-4, 4, 100)


μ = [0,0]
Σ = [2 0.0;
     0.0 2.0]
zs = [pdf(MvNormal(μ, Σ),[x,y]) for x in xs, y in ys]

contour!(ax2, xs, ys, zs)
xlims!(ax1,-4,4)
limits!(ax2,-4,4,-4,4)
ylims!(ax3,-4,4)
hideydecorations!(ax3, ticks = false, grid = false)
hidexdecorations!(ax1, ticks = false, grid = false)

fig[1,1] = ax2
fig[1,2] = ax3
fig[0,1] = ax1

colsize!(fig.layout, 1, Relative(2/3))
rowsize!(fig.layout, 1, Relative(1/3))
colgap!(fig.layout, 10)
rowgap!(fig.layout, 10)
Label(fig[0, :], text = "Acomplamento de Variáveis Independentes", textsize = 20)
save("acomplamento1.pdf", fig)
fig
```
