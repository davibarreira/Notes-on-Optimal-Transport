```julia
using CairoMakie
```

```julia
size_inches = (10, 8)
size_pt = 72 .* size_inches
```

```julia
x = y =  LinRange(-0.5, 1.2, 51)
z = x.^2 .+ (y'.^2)*0.2
zmin, zmax = minimum(z), maximum(z)
cmap = :viridis
fig = Figure(resolution = size_pt, fontsize = 12)
ax = Axis3(fig,
    aspect = :data, perspectiveness = 0.0, elevation = /9,
    zlabel = "F(x,y)",
#     xzpanelcolor= (:black, 0.75), yzpanelcolor= (:black,0.75),
    zgridcolor = :grey, ygridcolor = :grey,xgridcolor = :grey)
surface!(ax, x, y, z, colormap = cmap, colorrange = (zmin, zmax))
xm, ym, zm = minimum(ax.finallimits[])
contour!(ax, x, y, z, levels = 20, colormap = cmap, linewidth = 2,
    colorrange=(zmin, zmax), transformation = (:xy, zm))
wireframe!(ax, x, y, z, overdraw = false, transparency = true,
    color = (:black, 0.1))

t = 0:0.1:3
# lines!(exp.(-2 .*t), exp.(-2 .*t), t/4, color = t/4, linewidth = 4,
#     colormap = :plasma)
# lines!(1.5 .* exp.(-2 .*t), exp.(-0.4 .*t), (1.5 .* exp.(-2 .*t)).^2 .+ 0.2 .*(exp.(-0.4 .*t)).^2 .+ 0.1, color=:red, linewidth = 4)
meshscatter!(1.0 .* exp.(-2 .*t), exp.(-0.4 .*t), (1.0 .* exp.(-2 .*t)).^2 .+ 0.2 .*(exp.(-0.4 .*t)).^2 .+ 0.01,
    markersize=0.03, color=:red, opacity=0)
fig[1,1] = ax
fig
```

```julia
save("../Notes-Portugues/Figures/fluxogradiente.pdf", fig, pt_per_unit = 1)
```
